```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as shc
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

# **P1:**

```python
data = pd.read_csv('diabetes.csv', usecols=['Glucose', 'BloodPressure'])
data = data.head(20)
data
```

```python
# ploting before clustering

plt.figure(figsize=(12,5))
plt.scatter(data['Glucose'], data['BloodPressure'])
plt.xlabel('Glucose')
plt.ylabel('BloodPressure')
plt.show()
```

```python
# dendogram:

plt.figure(figsize=(14, 7))  
plt.title("Dendrograms")  
dend = shc.dendrogram(shc.linkage(data, method='single'))
plt.show()
```

```python
# Draw the line:

plt.figure(figsize=(14, 7))  
plt.title("Dendrograms")  
dend = shc.dendrogram(shc.linkage(data, method='single'))
plt.axhline(y=17, color='r', linestyle='--')
plt.show()
```

```python
cluster = AgglomerativeClustering(n_clusters=9, affinity='euclidean', linkage='single')
y_cluster = cluster.fit_predict(data)
y_cluster
```

```python
# ploting the clusters:

plt.figure(figsize=(10, 5))  
plt.scatter(data['Glucose'], data['BloodPressure'], c = cluster.labels_)
plt.show()
```

# **P4:**

**To create the cluster we need to scaling the data, so that distances among the data points/ individual clusters are minimum.**

Basically It depends on the data.

```python
data = pd.read_csv('Mall_Customers.csv', usecols=['Annual Income (k$)', 'Spending Score (1-100)'])
data.head()
```

```python
plt.figure(figsize=(10, 5))  
plt.scatter(data['Annual Income (k$)'], data['Spending Score (1-100)'])
plt.show()
```

```python
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)
scaled_data
```

```python
scaled_data = pd.DataFrame(scaled_data, columns=['Annual Income (k$)', 'Spending Score (1-100)'])
scaled_data.head()
```

```python
plt.figure(figsize=(10, 5))  
plt.scatter(scaled_data['Annual Income (k$)'], scaled_data['Spending Score (1-100)'])
plt.show()
```

```python
# dendogram:

plt.figure(figsize=(14, 7))  
plt.title("Dendrograms")  
dend = shc.dendrogram(shc.linkage(scaled_data, method='ward'))
plt.show()
```

```python
# Draw the line:

plt.figure(figsize=(14, 7))  
plt.title("Dendrograms")  
dend = shc.dendrogram(shc.linkage(scaled_data, method='ward'))
plt.axhline(y=8, color='r', linestyle='--')
plt.show()
```

```python
cluster = AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='ward')
y_cluster = cluster.fit_predict(scaled_data)
y_cluster
```

```python
# ploting the clusters:

plt.figure(figsize=(10, 5))  
plt.scatter(scaled_data['Annual Income (k$)'], scaled_data['Spending Score (1-100)'], c = cluster.labels_)
plt.show()
```
